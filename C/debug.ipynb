```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data_path = '../python/data/wedgeData_v3_128.txt'
```

```python
# read the first line of data
with open(data_path, 'r') as f:
    data_first_line = f.readline()
    
data_first_line_split = data_first_line[1:-2].split("),(")
```

```python
# number of chars of the first line
num_chars = len(data_first_line)
print("Number of chars in the first line: ", num_chars)
```

```python
# format: layer, radius, angle, z
layers = []
radii = []
angles = []
zs = []
for data in data_first_line_split:
    data_split = data.split(",")
    layers.append(int(data_split[0]))
    radii.append(float(data_split[1]))
    angles.append(float(data_split[2]))
    zs.append(float(data_split[3]))
```

```python
points = list(zip(layers, radii, angles, zs))
```

```python
# plot distribution of layers, radii, angles, zs
plt.figure(figsize=(12, 8))
plt.subplot(221)
plt.hist(layers, bins=128)
plt.title("Distribution of layers")
plt.xlabel("Layer")
plt.ylabel("Frequency")
plt.grid()
plt.subplot(222)
plt.hist(radii, bins=128)
plt.title("Distribution of radii")
plt.xlabel("Radius")
plt.ylabel("Frequency")
plt.grid()
plt.subplot(223)
plt.hist(angles, bins=128)
plt.title("Distribution of angles")
plt.xlabel("Angle")
plt.ylabel("Frequency")
plt.grid()
plt.subplot(224)
plt.hist(zs, bins=128)
plt.title("Distribution of zs")
plt.xlabel("z")
plt.ylabel("Frequency")
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# plot scatter plot using (layer, z)
plt.figure(figsize=(8, 6), dpi=100)
plt.scatter(zs, layers, s=20)
plt.title("Scatter plot of (z, layer)")
plt.xlabel("Layer")
plt.ylabel("z")
plt.grid()
plt.show()
```
